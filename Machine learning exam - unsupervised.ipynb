```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(card)

scaled_data = pd.DataFrame(scaled_data, columns=card.columns)

# Print the scaled data
card_s=card
```

```python
# for question b

Q1 = scaled_data.quantile(0.25)
Q3 = scaled_data.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = scaled_data[~((scaled_data <= upper_bound) & (scaled_data >= lower_bound))].index.tolist()

scaled_data.drop(outliers, inplace=True)

card.head()
```

Q3. Perform dimensionality reduction using PCA such that the 95% of the variance is explained

```python
# for question 3
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
pca.fit(card)
reduced_data = pca.transform(card)
print(reduced_data)

```

```python
import pandas as pd

card_reduced = pd.DataFrame(reduced_data)

card_reduced.columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'TENURE']

print(card_reduced.head())
```

Q4. Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curve

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(card_reduced)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
optimum_k = 3

kmeans = KMeans(n_clusters=optimum_k, random_state=0)
kmeans.fit(card_reduced)

card_reduced['Cluster'] = kmeans.labels_

print(card_reduced.head())
```

Q5. Find the optimum value of k for k-means clustering using the silhouette score method. Build a K-means clustering model and specify the number of observations in each cluster using a bar plot

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(card_reduced)
    silhouette_scores.append(silhouette_score(card_reduced, kmeans.labels_))

# Plot the silhouette scores
plt.plot(range(2, 11), silhouette_scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('Silhouette Score Method')
plt.show()

# Choose the optimum value of k
optimum_k = 4

# Perform k-means clustering with the optimum value of k
kmeans = KMeans(n_clusters=optimum_k, random_state=0)
kmeans.fit(card_reduced)

# Add the cluster labels to the dataframe
card_reduced['Cluster'] = kmeans.labels_

# Print the dataframe
print(card_reduced.head())

# Create a bar plot to show the number of observations in each cluster
sns.countplot(x='Cluster', data=card_reduced)
plt.show()
```
